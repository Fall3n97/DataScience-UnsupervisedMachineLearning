```python
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
from scipy.linalg import svd
from numpy import zeros
from numpy import diag
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.DataFrame(

  {

    "Matemáticas": [7.0, 7.5, 7.6, 5.0, 6.0, 7.8, 6.3, 7.9, 6.0, 6.8],

    "Ciencias": [6.5, 9.4, 9.2, 6.5, 6.0, 9.6, 6.4, 9.7, 6.0, 7.2],

    "Español": [9.2, 7.3, 8.0, 6.5, 7.8, 7.7, 8.2, 7.5, 6.5, 8.7],

    "Historia": [8.6, 7.0, 8.0, 7.0, 8.9, 8.0, 9.0, 8.0, 5.5, 9.0],

    "EdFísica": [8.0, 7.0, 7.5, 9.0, 7.3, 6.5, 7.2, 6.0, 8.7, 7.0],

  },

  index=["Lucía", "Pedro", "Inés", "Luis", "Andrés", "Ana", "Carlos", "José", "Sonia", "María"]

)

df
```

```python
labels = df.index
labels
```

```python
notes = df.reset_index(drop = True)
notes
```

```python
U, s, VT = svd(notes)    # Obtencion de matrices
Sigma = zeros((notes.shape[0], notes.shape[1]))   # Creacion de matriz de ceros de la misma dimensión que notes
Sigma[:notes.shape[1], :notes.shape[1]] = diag(s)
```

```python
print("Matriz U:")
print(U)
```

```python
print("Matriz Sigma:")
print(Sigma)
```

```python
print("Matriz VT:")
print(VT)
```

```python
# Reduccion a DOS Dimensiones
n_elements = 2

UReduced2 = U[:, :n_elements]
print("U Reducida: ")
print(UReduced2)
```

```python
SigmaReduced2 = Sigma[:n_elements, :n_elements]
print("Matriz Sigma Reducida: ")
print(SigmaReduced2)
```

```python
VTReduced2 = VT[:n_elements, :]
print("VT Reducida: ")
print(VTReduced2)
```

```python
NotesReduced2 = UReduced2.dot(SigmaReduced2.dot(VTReduced2))
print("Matriz notes Transformada: ")
print(NotesReduced2)
```

```python
T2 = UReduced2.dot(SigmaReduced2)
print('Matriz T (2 Dimensiones): ')
print(T2)
```

```python
df = pd.DataFrame(T2, columns=['a', 'b'])
df.plot(kind = "scatter", x = 'a', y = 'b')
plt.show
```

```python
df = pd.DataFrame(T2, columns=['a', 'b'])
x = df.iloc[:, 0]
y = df.iloc[:, 1]
x = x.to_numpy()
y = y.to_numpy()

fig, ax = plt.subplots()
ax.set_title("Mapa de observaciones")
ax.scatter(x,y)

for i, txt in enumerate(labels):
    ax.annotate(txt, (x[i], y[i]))
```

```python
# Reduccion a TRES Dimensiones
n_elements = 3

UReduced3 = U[:, :n_elements]

SigmaReduced3 = Sigma[:n_elements, :n_elements]

VTReduced3 = VT[:n_elements, :]

NotesReduced3 = UReduced3.dot(SigmaReduced3.dot(VTReduced3))
print("Matriz notes Transformada: ")
print(NotesReduced3)
```

```python
T3 = UReduced3.dot(SigmaReduced3)
print('Matriz T (3 Dimensiones): ')
print(T3)
```

```python
df3 = pd.DataFrame(T3, columns = ['a', 'b', 'c'])
df3
```

```python
ax3d = plt.figure().add_subplot(projection='3d')

arrayx = np.array(df3['a'])
arrayy = np.array(df3['b'])
arrayz = np.array(df3['c'])

ax3d.scatter(arrayx, arrayy, arrayz)

for x, y, z, label in zip(arrayx, arrayy, arrayz, labels):
    ax3d.text(x,y,z, label)

plt.title("Datos en 3D")
plt.show
```
