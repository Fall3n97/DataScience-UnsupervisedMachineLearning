```python
import os
import pandas as pd
os.chdir('./')

customer_data_file = 'Online Retail.xlsx'
datos = pd.read_excel(customer_data_file)
datos.head(10)
```

```python
# Eliminar todos lo valores nulos
datos = datos.dropna()
datos.info()
```

```python
# Delimitacion de Valores Positivos del campo 'Quantity' (Pues deseamos analizar lo realmente comprado)
datos_reales = datos[datos['Quantity'] > 0]
datos_reales.info()
```

```python
basket = (datos_reales.groupby(['InvoiceNo', 'Description']))['Quantity'].sum().unstack().reset_index().fillna(0).astype(int).set_index('InvoiceNo')
basket
```

```python
# Codificacion a Binario de la base previa para identificar si un articulo se compro o no
def codificar(x):
    if x <= 0:
        return 0
    if x > 1:
        return 1

basket_codificado = basket.map(codificar).fillna(0).astype(int)
basket_codificado
```

```python
# Filtrado de aquellas transacciones en donde se ha comprado mas de un articulo
basket_filtrado = basket_codificado[(basket_codificado > 0).sum(axis = 1) >= 2]
basket_filtrado
```

```python
# Aplicacion del algoritmo 'A priori' para localizar los articulos frecuentemente comprados en la base de datos
from mlxtend.frequent_patterns import apriori
```

```python
# Analisis 'a priori' con un soporte minimo del 3%
articulos_frecuentes = apriori(basket_filtrado, min_support = 0.025,
                              use_colnames = True).sort_values('support', ascending = False).reset_index(drop = True)
articulos_frecuentes['length'] = articulos_frecuentes['itemsets'].apply(lambda x:len(x))
articulos_frecuentes
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(articulos_frecuentes, metric = 'confidence',
                 min_threshold = 0.5).sort_values('confidence', ascending = False).reset_index(drop = True)
```

```python
association_rules(articulos_frecuentes, metric = 'lift',
                 min_threshold = 1.0).sort_values('lift', ascending = False).reset_index(drop = True)
```
